```
x = np.arange(1,5)
y = x**3
plt.plot([1,2,3,4],[1,4,9,16],"go",x,y,"r^")
plt.title("First Plot")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.show()
```

```
x = np.arange(1,5)
y = x**3
plt.subplot(1,2,1)
plt.plot([1,2,3,4],[1,4,9,16],"go")
plt.title("1st Subplot")

plt.subplot(1,2,2)
plt.plot(x,y,"r^")
plt.title("2nd Subplot")

plt.suptitle("My Sub-Plots")
plt.show()
```

```
x = np.arange(1,5)
y = x**3
plt.subplot(2,1,1)
plt.plot([1,2,3,4],[1,4,9,16],"go")
plt.title("1st Subplot")

plt.subplot(2,1,2)
plt.plot(x,y,"r^")
plt.title("2nd Subplot")

plt.suptitle("My Sub-Plot")
plt.show()
```

```
x = np.arange(1,5)
y = x**4
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(6,6))
ax[0,0].plot([1,2,3,4],[1,2,3,4],"y")
ax[0,0].set_title("Line")
ax[0,1].plot([1,2,3,4],[1,4,9,16],"go")
ax[0,1].set_title("Circle")
ax[1,0].plot([1,2,3,4],[1,4,9,16],"r^")
ax[1,0].set_title("Triangule")
ax[1,1].plot(x,y,"b*")
ax[1,1].set_title("Start")
```

```
divisions = ["Div-A","Div-B","Div-C","Div-D","Div-E"]
division_average_marks = [70, 82, 73, 65, 68]

plt.bar(divisions, division_average_marks, color = "blue")
plt.title("Bar Graph")
plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.show()
```

```
divisions = ["Div-A","Div-B","Div-C","Div-D","Div-E"]
division_average_marks = [70, 82, 73, 65, 68]

plt.barh(divisions, division_average_marks, color = "blue")
plt.title("Bar Graph")
plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.show()
```

```
divisions = ["Div-A","Div-B","Div-C","Div-D","Div-E"]
division_average_marks = [70, 82, 73, 65, 68]
boys_average_marks = [68, 67, 77, 62, 73]

index = np.arange(5)
width = 0.15

plt.bar(index, division_average_marks, width, color = "blue", label = 'Division Marks')
plt.bar(index + width, boys_average_marks, width, color = "green", label = 'Boys Marks')
plt.title("Horizontally Stacked Bar Graphs")

plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.xticks(index + width/2, divisions)

plt.legend(loc = 'best')
plt.show()
```

```
divisions = ["Div-A","Div-B","Div-C","Div-D","Div-E"]
division_average_marks = [70, 82, 73, 65, 68]
boys_average_marks = [68, 67, 71, 62, 65]

index = np.arange(5)
width = 0.15

plt.bar(index, division_average_marks, width, color = "blue", label = 'Division Marks')
plt.bar(index, boys_average_marks, width, color = "red", label = 'Boys Marks', bottom = boys_average_marks)
plt.title("Horizontally Stacked Bar Graphs")

plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.xticks(index + width/2, divisions)

plt.legend(loc = 'best')
plt.show()
```

```
firms = ["Firm A","Firm B", "Firm C" , "Firm D", "Firm E"]
market_share = [20,25,15,10,20]
Explode = [0,0.1,0,0,0]
plt.pie(market_share,explode=Explode,labels = firms, shadow=True, startangle = 45)
plt.axis('equal')
plt.legend(title = "List of firms", loc = 'lower right')
plt.show()


```

```
x = np.random.randn(1000)

plt.title("Histogram")
plt.xlabel("Random Data")
plt.ylabel("Frequency")
plt.hist(x,10)
plt.show()
```

```
height = np.array([157,170,180,189,167,178,189,190,199,200])
weight = np.array([57,70,80,89,67,78,89,90,99,100])

plt.xlim(140,210)
plt.ylim(50,110)
plt.scatter(height,weight)
plt.title("Scatter Plot")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```
